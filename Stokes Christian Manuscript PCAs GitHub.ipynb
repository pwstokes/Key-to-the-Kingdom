```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

## CSV Import and Binarization

```python
def binarizeData(csv):
    #Read .csv file
    data = pd.read_csv(csv)

    #Binarize the data
    data = pd.get_dummies(data.drop(['Biblical Citation', 'Variants'],1).dropna().transpose())
    
    return data
```

## PCA

```python
pca = PCA(n_components=2)
```

## Figure set-up

```python
def plotFigure(output,data,colors):
    maxvalues = output.max(0)
    minvalues = output.min(0)

    plt.rcParams['font.size'] = 12
    fig, ax = plt.subplots(1, figsize=(15, 15))
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(r'1st principal component $\rightarrow$')
    plt.ylabel(r'2nd principal component $\rightarrow$')

    ax.scatter(*output.T, s = 100, c = colors)
    for i, r in enumerate(data.index[:]):
        ax.text(output[i,0], output[i,1], r)
```

## PCA Lexical Variation in "Kingdom of Heaven/God"

### Figure 1

```python
colors = ['grey','blue', 'blue', 'blue', 'red', 'orange', 'yellow', 'green', 'green', 'green', 'green', 'green', 'green', 'green', 'green', 'green', 'aqua', 'purple', 'purple', 'purple', 'purple', 'purple']
PCA1data = binarizeData('1 Kingdom of God:Heaven ALL.csv')
PCA1 = pca.fit_transform(PCA1data)
plotFigure(PCA1,PCA1data,colors)
print(pca.explained_variance_ratio_)
```
